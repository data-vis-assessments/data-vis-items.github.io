```python
##BRBF
```

```python
brbf = pd.read_csv("/Users/averyyue/Desktop/data-vis-review/brbf_questions.csv")
```

```python
columns_to_check = [f'mc{i}' for i in range(9, 16)]  # mc9 to mc15
count = brbf[columns_to_check].dropna().shape[0]
print(f"Number of rows with data in all mc9-mc15 columns: {count}")
```

```python
column_mapping = {f'mc{i}': f'answer_{i}' for i in range(1, 16)}
brbf = brbf.rename(columns=column_mapping)
```

```python
brbf['Unnamed: 0'] = [f'BRBF_{i}' for i in range(1, len(brbf) + 1)]
```

```python
brbf = brbf.rename(columns={'Unnamed: 0':'item_ids', 'task_category' : 'task_types', 'graph_type':'graph_types', 'image_link':'graph_url'})
```

```python
def rename_graphs(row):
    graph = row['graph_types']
    print(graph)
    if pd.isna(graph):
        row['graph_types_ctl'] = np.nan
        return row
    if 'table' in graph:
        row['graph_types_ctl'] = 'Table'
    if 'bg' in graph:
        row['graph_types_ctl'] = 'Bar chart'
    if 'lg' in graph:
        row['graph_types_ctl'] = 'Line graph'
    if 'sp' in graph:
        row['graph_types_ctl'] = 'Scatter plot'
    return row
brbf = brbf.apply(rename_graphs, axis = 1)
```

```python
cols = ['item_ids', 'task_types', 'graph_types', 'graph_url', 'graph_types_ctl', 'question']
lst = [f'answer_{i}' for i in range(1, 16)]
total = cols + lst
brbf_cleaned = brbf[total].copy()
```

```python
brbf_cleaned
```

```python
for i in range(len(brbf_cleaned)):
    brbf_cleaned.loc[i, 'task_types_ctl'] = brbf_cleaned.iloc[i]['task_types']
    brbf_cleaned.loc[i, 'test_name'] = 'BRBF'
    brbf_cleaned.loc[i, 'open_answer'] = 'multiple-choice'
```

```python
brbf_cleaned.to_csv('/Users/averyyue/Downloads/data-vis-dashboard/public/BRBF.csv')
```

```python
##GGR
```

```python
ggr = pd.read_csv("/Users/averyyue/Desktop/data-vis-review/ggr_questions.csv")
```

```python
def rename_inputs(value):
    if pd.isna(value):
        return np.nan
    elif value == 'open_ended_question':
        return 'open-answer'
    elif value == 'multiple_choice_question':
        return 'multiple-choice'
ggr['open_answer'] = ggr['text_input_type'].apply(rename_inputs)
```

```python
ggr['Unnamed: 0'] = [f'GGR_{i}' for i in range(1, len(ggr) + 1)]
column_mapping = {f'mc{i}': f'answer_{i}' for i in range(1, 16)}
ggr = ggr.rename(columns=column_mapping)
ggr = ggr.rename(columns={'task_category':'task_types', 'image_link':'graph_url', 'Unnamed: 0' : 'item_ids', 'chart_type':'graph_types'})
```

```python
ggr_cleaned = ggr[['task_types', 'graph_url', 'item_ids', 'graph_types', 'answer_1', 'answer_2', 'answer_3', 'answer_4', 'open_answer', 'question']].copy()
ggr_cleaned.head()
```

```python
def standard_format(row):
    graph = row['graph_types']
    lower = graph[0] + graph[1:].lower()
    print(lower)
    row['graph_types_ctl'] = lower
    return row
ggr_cleaned = ggr_cleaned.apply(standard_format, axis=1)
```

```python
for i in range(len(ggr_cleaned)):
    ggr_cleaned.loc[i, 'task_types_ctl'] = ggr_cleaned.iloc[i]['task_types']
    ggr_cleaned.loc[i, 'test_name'] = 'GGR'
```

```python
ggr_cleaned.to_csv('/Users/averyyue/Downloads/data-vis-dashboard/public/GGR.csv')
```
